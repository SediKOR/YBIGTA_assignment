### 5) Feature Engineering

feature에 대한 전처리를 하기 이전에 결측치가 있는 column을 모두 제거했는데, 다시 초기 상태로 되돌리고 진행하겠습니다.

```python
df = pd.read_csv('drive/MyDrive/airbnb_listings.csv', index_col=0)
```

- #### <font color='orange'> Q12. 결측치에 대한 처리를 하고 싶습니다.
#### <font color='orange'> 주어진 모든 column들에 대하여 결측치가 25%이상인 경우에는 column을 drop하고 10%이상 25%미만인 경우에는 결측치가 있는 row만 drop하고, 10%미만인 경우에는 중앙값으로 대체해주세요.

```python
#Q12

row_drop=[]

for i in df.columns:
   missing_rate = (df[i].isnull().sum()) / len(df[i])
   if(missing_rate >= 0.25):
    df.drop(i, axis=1, inplace=True)

   elif(missing_rate >= 0.10):
    row_drop.append(i)

   elif(missing_rate < 0.10 and df[i].dtypes != object):
    median = df[i].median()
    df[i].fillna(median, inplace=True)

df.dropna(subset=row_drop, inplace=True)
```

- #### <font color='orange'> Q13. 앞서 진행한 EDA를 바탕으로 FE를 진행하고자 합니다..
#### <font color='orange'> 총 3개의 FE를 진행하시고, 해당 FE를 진행한 이유에 대해 **간단히** 작성해주세요.


```python
#Q13-1) bed_type에 대해 one-hot encoding을 진행. 한 눈에 방 컨디션을 비교 편하기 때문

pd.get_dummies(df['bed_type'])
```

```python
#Q13-2) review score rating에 자연어 평가 조건문 추가. 점수에 따라 추천 리뷰 글자를 볼 수 있음

rating = []
for r in df['review_scores_rating']:
    if (r >= 90):
        rating.append('Great')
    elif (r >= 50):
        rating.append('Good')
    else:
        rating.append('BAD')

df['rating'] = rating

display(df)
```

```python
#Q13-3) 만든 rating 자연어 평가에 대해 one-hot encoding을 통해 한 눈에 파악하도록 제작

pd.get_dummies(df['rating'])
```
